# The Hello World of Tensor Flow!
## y=2x-1
### We will give x's and y's to a model and it will come up with the above equation

Imports

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

Create the Neural Network. It has 1 layer, which in turn has 1 neuron. The input shape is 1 value.

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Compile the Neural Network. We need 2 functions
## A loss function
This measures the guessed ansers against the known correct answers and measures how well/poorly it did.
## An optimizer
The optimizer makes another guess. Based on how well the loss function went, it tries to minimize the loss.



```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Providing the data. Here are the first 6 values for y=2x-1. The data is stored in numpy arrays

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

Training the neural network. Going through the above 'loop'. It will make a guess. Measure the loss. Optimize. Make another guess. It will conduct this cycle for the amount of epochs specified. Below is 500 epochs

```
model.fit(xs, ys, epochs=500)
```

The model has been trained and learned the relationship between x and y. We can feed in an x that the model has not seen before, and it will predict y.

```
model.predict([10.0])
```
